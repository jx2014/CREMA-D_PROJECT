# Logistic Regression

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
data = np.load(r'crema_data.npz')

X = data['features']   
y = data['emotions']   
filenames = data['filenames'] 
```

```python
y
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
for emotion, code in zip(le.classes_, range(len(le.classes_))):
    print(f"{emotion} : {code}")
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=int(time.time()), stratify=y_encoded
)
```

```python
print(f"\nTrain size: {len(X_train)}")
print(f"Test size: {len(X_test)}")

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Try different regularization strengths
C_values = [0.01, 0.1, 1, 10, 100]
best_accuracy = 0
best_model = None
best_C = None
```

```python
for C in C_values:
    model = LogisticRegression(C=C, max_iter=1000, random_state=int(time.time()))
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    print(f"C={C:5} | Accuracy: {acc:.4f}")
    
    if acc > best_accuracy:
        best_accuracy = acc
        best_model = model
        best_C = C

print(f"\n✓ Best C={best_C} with accuracy: {best_accuracy:.4f}")
```
